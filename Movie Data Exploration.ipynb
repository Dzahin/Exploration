```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
#import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\PaperPlane\Downloads\movies.csv')
```

```python
df.head()
```

```python
#check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#remove missing data
df.dropna(inplace=True)
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
#change data type - remove decimal point
df['budget']=df['budget'].astype('int64')
df['votes']=df['votes'].astype('int64')
df['gross']=df['gross'].astype('int64')
df.head()
```

```python
df['corrected_relesed']=df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
df.head()
```

```python
df.sort_values(by=['gross'], ascending= False)
df.head()
```

```python
#drop duplicates
df.drop_duplicates()
df.head()
```

```python
#Looking for outlier in gross
df.boxplot(column=['gross'])
```

```python
# scatter plot budget vs gross
plt.scatter(x=df['budget'], y=df['gross'],alpha=0.5)
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
#plot budget vs gross using sns reg

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'green'}, line_kws={'color':'red'})
```

```python
#correlation matrix
df.corr(method='pearson')
```

```python
#visualise correlation
cm = df.corr(method='pearson')
sns.heatmap(cm,annot = True)
plt.title('Movie Correlation Matrix')
plt.xlabel('Movie Attribute')
plt.ylabel('Movie Attribute')

plt.show()
```

```python
#changing string into index

dm=df

for col in dm.columns:
    if dm[col].dtype == 'object':
        dm[col] = df[col].astype('category').cat.codes

dm.head()
```

```python
#visualising all movie att.

cm = dm.corr(method='pearson')
sns.heatmap(cm,annot = True)
plt.title('Movie Correlation Matrix')
plt.xlabel('Movie Attribute')
plt.ylabel('Movie Attribute')

plt.show()
```
